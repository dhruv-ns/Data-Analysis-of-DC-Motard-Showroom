```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('C:\\Users\\Dhruv\\Documents\\Model_specs-1.csv')
df1 = pd.read_csv('C:\\Users\\Dhruv\\Documents\\Customer.csv')

while True:
    print('-------------')
    print('| DC MOTARD |')
    print('-------------', '\n')
    print('  Main Menu  ')
    print('-------------')
    print('''| 1.Display Data       |
           \n| 2.Accessing Data     |
           \n| 3.Data Analysis      |
           \n| 4.Data Visualisation |
           \n| 5.Exit               |''')
    print('--------------------------')
    print()
    ch = int(input("Enter your choice:"))
    if ch == 1:
        print('--------------------')
        print('|   Display Data   |')
        print('''| 1.Displaying full data        |
               \n| 2.Displaying customer details |
               \n| 3.Displaying first n rows     |
               \n| 4.Displaying last n rows      |
               \n| 5.Exit                        |''')
        print('-----------------------------------')
        print()
        dis = int(input('Enter your choice:'))
        if dis == 1:
            print('---{ DISPLAYING FULL DATA }---')
            print()
            print(df)
        elif dis == 2:
            print('---{ DISPLAYING CUSTOMER DETAILS }---')
            print()
            print(df1)
        elif dis == 3:
            print('---{ FIRST n ROWS }---')
            print()
            f = int(input('Enter the number the number of rows to be displayed:'))
            print()
            print(df.head(f))
        elif dis == 4:
            print('---{ LAST n ROWS }---')
            print()
            s = int(input('Enter the number of rows to be displayed:'))
            print(df.tail(s))
        elif dis == 5:
            pass
        else:
            print('INVALID CHOICE')
    elif ch == 2:
        print('----------------------')
        print('|   Accessing data   |')
        print('''| 1.Subsets         |
               \n| 2.Individual rows |
               \n| 3.Multiple rows   |
               \n| 4.Exit            |''')
        print('-----------------------')
        print()
        a = int(input('Enter your choice:'))
        if a == 1:
            print(df.columns)
            print()
            s1 = int(input('Enter the starting row index:'))
            s2 = int(input('Enter the ending row index:'))
            s3 = input('Enter the starting column:')
            s4 = input('Enter the ending column:')
            s3 = s3.capitalize()
            s4 = s4.capitalize()
            print()
            print('---------------')
            print('|   SUBSETS   |')
            print('---------------')
            print(df.loc[s1:s2,s3:s4])
        elif a == 2:
            print(df.columns)
            print()
            i1 = int(input('Enter the row index to be accessed:'))
            print()
            print('---------------------------------')
            print('|   ACCESSING INDIVIDUAL ROWS   |')
            print('---------------------------------')
            print(df.loc[i1])
        elif a == 3:
            print(df.columns)
            print()
            i1 = int(input('Enter the starting row index:'))
            print()
            i2 = int(input('Enter the ending row index:'))
            print()
            print('-------------------------------')
            print('|   ACCESSING MULTIPLE ROWS   |')
            print('-------------------------------')
            print(df.loc[i1:i2])
        elif a == 4:
            pass
        else:
            print('INVALID CHOICE')
    elif ch == 3:
        print('---------------------')
        print('|   DATA ANALISIS   |')
        print('---------------------')
        print('''| 1.Minimum value              |
               \n| 2.Maximum value              |
               \n| 3.Information regarding data |
               \n| 4.Shape                      |                      |
               \n| 5.Sort                       |
               \n| 6.Group by                   |
               \n| 7.Exit                      |''')
        print('----------------------------------')
        print()
        an = int(input('Enter your choice:'))
        print()
        if an == 1:
            print(df.columns)
            print()
            print('TO DISPLAY THE MINIMUM VALUE')
            print()
            m = input('Enter the column name:')
            m = m.capitalize()
            print()
            print('Min value:', df[m].min())
        elif an == 2:
            print(df.columns)
            print()
            print('TO DISPLAY THE MAXIMUM VALUE')
            print()
            m = input('Enter the columns name:')
            m = m.capitalize()
            print()
            print('Max value:', df[m].max())
        elif an == 3:
            print('THE INFORMATION REGARDING DATA')
            print()
            print(df.info())
        elif an == 4:
            print('SHAPE OF THE DATA')
            print()
            print(df.shape)
        elif an == 5:
            print('------------')
            print('|   SORT   |')
            print('------------')
            print('''| 1.Sort in ascending order  |
                   \n| 2.Sort in descending order |
                   \n| 3.Exit                     |''')
            print('--------------------------------')
            print()
            z = int(input('Enter your choice:'))
            if z == 1:
                print(df.columns)
                print()
                print('FOR SORTING IN ASCENDING ORDER')
                print()
                s = input('Enter the column name:')
                s = s.capitalize()
                print()
                print(df.sort_values(s,inplace=False))
                print()
                print('SORTED IN ASCENDING ORDER')
            elif z == 2:
                print(df.columns)
                print()
                print('FOR SORTING IN DESCENDING ORDER')
                print()
                s = input('Enter the column name:')
                s = s.capitalize()
                print(df.sort_values(s,inplace=False,ascending=False))
                print()
                print('SORTED IN DESCENDING ORDER')
            elif z == 3:
                pass
            else:
                print('INVALID VALUE')
        elif an == 6:
            print('----------------')
            print('|   GROUP BY   |')
            print('----------------')
            print(df.columns)
            print()
            o1 = input('Enter the first column name:')
            o1 = o1.capitalize()
            print()
            o2 = input('Enter the second column name:')
            o2 = o2.capitalize()
            print(df.groupby(o1)[o2].count())
            print()
        elif an == 7:
            pass
        else:
            print('INVALID CHOICE')
    elif ch == 4:
        print('--------------')
        print('|   DATA VISUALISATION   |')
        print('--------------------------')
        print('''| 1.Sales in 4 years 2020,2021,2022,2023 |
               \n| 2.Sales of year 2020                   |
               \n| 3.Sales of year 2021                   |
               \n| 4.Sales of year 2022                   |
               \n| 5.Sales of year 2023                   |
               \n| 6.Total sales                          |
               \n| 7.Exit                                 |''')
        print('--------------------------------------------')
        print()
        v = int(input('Enter your choice:'))
        print()
        if v == 1:
            print('--------------------------------------------')
            print('|   SALES IN 4 YEARS 2020,2021,2022,2023   |')
            print('--------------------------------------------')
            print('''| 1.Bar chart  |
                   \n| 2.Exit       |''')
            print('------------------')
            print()
            ap = int(input('Enter your choice:'))
            if ap == 1:
                import pandas as pd
                import matplotlib.pyplot as plt
                plotdata=pd.DataFrame(
                    {'2020':[5,4,8,10],
                    '2021':[25,18,19,11],
                    '2022':[11,14,20,23],
                    '2023':[10,14,19,33]},
                    index=['BMW','Ducati','Kawasaki','Harley davidson'])
                plotdata.plot(kind='bar')
                plt.title('sales of bikes')
                plt.xlabel('brands')
                plt.ylabel('bike sale')
                plt.show()
            elif ap == 2:
                pass
            else:
                print('INVALID CHOICE')
        elif v == 2:
            print('--------------------------')
            print('|   Sales of year 2020   |')
            print('--------------------------')
            print()
            import matplotlib.pyplot as plt
            Brand=['BMW','Ducati', 'Kawasaki', 'Harley davidson']
            unit=[25,33,46,52]
            plt.plot(Brand,unit)
            plt.xlabel('Brand')
            plt.ylabel('unit')
            plt.title('sales of bikes in 2020')
            plt.grid(True)
            plt.yticks(unit)
            plt.show()
        elif v == 3:
            print('--------------------------')
            print('|   Sales of year 2021   |')
            print('--------------------------')
            print()
            import matplotlib.pyplot as plt
            Brand=['BMW','Ducati', 'Kawasaki', 'Harley davidson']
            unit=[35,23,26,42]
            plt.plot(Brand,unit)
            plt.xlabel('Brand')
            plt.ylabel('unit')
            plt.title('sales of bikes in 2021')
            plt.grid(True)
            plt.yticks(unit)
            plt.show()
        elif v == 4:
            print('--------------------------')
            print('|   Sales of year 2022   |')
            print('--------------------------')
            print()
            import matplotlib.pyplot as plt
            Brand=['BMW','Ducati', 'Kawasaki', 'Harley davidson']
            unit=[50,53,46,56]
            plt.plot(Brand,unit)
            plt.xlabel('Brand')
            plt.ylabel('unit')
            plt.title('sales of bikes in 2022')
            plt.grid(True)
            plt.yticks(unit)
            plt.show()
        elif v == 5:
            print('--------------------------')
            print('|   Sales of year 2023   |')
            print('--------------------------')
            print()
            import matplotlib.pyplot as plt
            Brand=['BMW','Ducati', 'Kawasaki', 'Harley davidson']
            unit=[25,43,46,55]
            plt.plot(Brand,unit)
            plt.xlabel('Brand')
            plt.ylabel('unit')
            plt.title('sales of bikes in 2023')
            plt.grid(True)
            plt.yticks(unit)
            plt.show()
        elif v == 6:
            print('-------------------')
            print('|   TOTAL SALES   |')
            print('-------------------')
            print()
            import pandas as pd
            import matplotlib.pyplot as plt
            plotdata=pd.DataFrame(
                {'2020': [10,13,12,14], '2021':[12,14,15,17], '2022':[15,17,15,16], '2023':[16,14,19,18]}, index=['BMW', 'Ducati', 'Kawasaki', 'Harley davidson'])
            plotdata.plot(kind='bar', stacked=True)
            plt.title('sales of bikes')
            plt.xlabel('Brands')
            plt.ylabel('Units sell')
            plt.show()
        elif v == 7:
            pass
        else:
            print('INVALID CHOICE')
    elif ch == 5:
        break
    else:
        print('INVALID CHOICE')
```
